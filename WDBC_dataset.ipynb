```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.countplot(x='target', data=df)
plt.title('Distribution of Malignant vs Benign Tumors')
plt.xlabel('Diagnosis')
plt.ylabel('FNA Breast Mass Samples')
palette='viridis'
plt.show()
```

```python
sns.boxplot(x='target', y='mean radius', data=df)  # Changed 'radius_mean' to 'mean radius'
plt.title('Radius Mean Distribution by Diagnosis')
plt.show()

sns.histplot(df[df['target'] == 0]['mean radius'], kde=True, label='Benign', color='blue') # Changed 'radius_mean' to 'mean radius'
sns.histplot(df[df['target'] == 1]['mean radius'], kde=True, label='Malignant', color='red') # Changed 'radius_mean' to 'mean radius'
plt.title('Radius Mean Distribution')
plt.legend()
plt.show()
```

```python
sns.scatterplot(x='mean radius', y='mean area', hue='target', data=df, palette='coolwarm')
plt.title('Radius Mean vs. Area Mean by Diagnosis')
plt.show()
```

```python
sns.boxplot(x='target', y='mean smoothness', data=df)  # Changed 'smoothness_mean' to 'mean smoothness'
plt.title('Smoothness Mean Distribution by Diagnosis')
plt.show()
```

```python
sns.violinplot(x='target', y='mean compactness', data=df)  # Changed 'compactness_mean' to 'mean compactness'
plt.title('Compactness Mean Distribution by Diagnosis')
plt.show()
```
